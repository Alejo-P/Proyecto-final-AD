```python
import pandas as pd
```

```python
url='https://es.wikipedia.org/wiki/Elecciones_en_Ecuador'
```

```python
elecciones = pd.read_html(url)
```

```python
elec1 = elecciones[0]
```

```python
#Revisión de nulos
elec1.isnull().sum()
```

```python
generales=elec1.dropna()
```

```python
generales.isnull().sum()
```

```python
generales.to_csv('eleccionesGenerales2023.csv')
```
